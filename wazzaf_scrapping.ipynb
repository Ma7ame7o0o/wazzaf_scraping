# Web_Scraping

```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
from openpyxl import load_workbook
from openpyxl.styles import Font, Alignment, PatternFill
```

```python
# init empty lists to store info
job_titles = []
job_locations = []
job_companies = []
timeeons = []
```

```python
x = 0
while x <= 2:
    url = f"https://wuzzuf.net/search/jobs/?q=data+analysis&a=navbl&start={x}"
    response = requests.get(url)
    src = response.content
    print(src)
    x += 1

```

```python
# Parsing
soup = BeautifulSoup(src, 'html.parser')
soup
```

```python
#lxml best C ,fast 
soupp = BeautifulSoup(src, "lxml")
soupp
```

```python
# Findalll
titles = soup.find_all('h2', {'class': 'css-193uk2c'})
companies = soup.find_all('a', {'class': 'css-ipsyv7'})
locations = soup.find_all('span', {'class': 'css-16x61xq'})
times = soup.find_all('span', {'class': 'eoyjyou0'})
```

```python
# Loop to fill my listsssss
for i in range(len(titles)):
    job_titles.append(titles[i].text.strip())
    job_companies.append(companies[i].text.strip())
    job_locations.append(locations[i].text.strip())
    timeeons.append(times[i].text.strip())
```

```python
# Print the extracted data to check the output
print("Job Titles:", job_titles)
print("Companies:", job_companies)
print("Locations:", job_locations)
print("Posted Time:", timeeons)
```

```python
# make ourrr Dictionary
data = {
    "Job Title": job_titles,
    "Company": job_companies,
    "Location": job_locations,
    "Posted Time": timeeons
}
```

```python
#use pd to make a data frame
df = pd.DataFrame(data)
```

```python
#save on excel file
df.to_excel("jobs.xlsx", index=False)
```

```python
file_path = "jobs.xlsx"
wb = load_workbook(file_path)
ws = wb.active

# Format header row
for cell in ws[1]:
    cell.font = Font(bold=True, color="FFFFFF")
    cell.fill = PatternFill(start_color="4F81BD", fill_type="solid")
    cell.alignment = Alignment(horizontal="center")

# Adjust column width automatically
for column in ws.columns:
    max_length = 0
    column_letter = column[0].column_letter
    for cell in column:
        try:
            if cell.value:
                max_length = max(max_length, len(str(cell.value)))
        except:
            pass
    ws.column_dimensions[column_letter].width = max_length + 3

# Save changes
wb.save(file_path)

print(f"Formatted file saved at: {file_path}")
```
